# Moving Average of stock price

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import pandas.stats.moments as pdsm
```

```python
HSBC = pd.read_csv('0005HK.csv', header= None)
HSBC.columns = ['date','price']
date = HSBC['date']
price = HSBC['price']

# put date into date format for plotting
# create a string representing the time under the contorl of an explicit format
x = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in HSBC['date']]
```

```python
print(x)
```

```python
# extract the part from 2013
dprice = price[0:].values
```

```python
dprice
```

```python
# Calculate moving average of 50 days
#ma = pdsm.rolling_mean(dprice,50)
ma = pdsm.Series(dprice).rolling(50).mean()
```

```python
ma.head(100)
```

```python
# plot graphs
plt.plot(x[49:],ma[49:])
plt.plot(x[49:],price[49:])

# format x-axis automatically
plt.gcf().autofmt_xdate()

# show the graph
plt.show()
```
